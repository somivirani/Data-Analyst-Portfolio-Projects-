```python
# Libraries to extract, process, and organize website information into tables.
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO
```

```python
# Fetches and parses Wikipedia's Economy of India data page for web scraping.
```

```python
url = "https://en.wikipedia.org/wiki/Economy_of_India#Data"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser') 
```

```python
# Extracts all 'table' HTML elements with the 'wikitable' CSS class, for data retrieval.
```

```python
tables = soup.find_all('table', {'class': 'wikitable'})
tables   
```

```python
# Table 1: Key Economic Indicators
df_gdp = pd.read_html(StringIO(str(tables[0])))[0]
df_gdp
```

```python
# Table 2: GDP by Sector
df_sector = pd.read_html(StringIO(str(tables[1])))[0]
df_sector    
```

```python
 # Table 3: Government Finances
df_govt_finance = pd.read_html(StringIO(str(tables[3])))[0]
df_govt_finance   
```

```python
# Imports common data analysis, visualization, and cleaning libraries: pandas, numpy, matplotlib, seaborn, and sets matplotlib to display inline.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#deleting null values
```

```python
df_gdp=df_gdp.dropna()
```

```python
df_gdp.isnull().sum().sum()
```

```python
# Cleaning GDP data:
# 1. Remove columns with all missing values.
# 2. Rename columns for clarity.
```

```python
df_gdp = df_gdp.dropna(axis=1, how='all')
df_gdp.columns = ['Year', 'GDP_Nominal', 'GDP_Nominal_USD',
                   'GDP_Real', 'GDP_Per_Capita',
                   'GDP_Per_Capita_USD', 'GDP_Growth',
                   'Inflation', 'Govt_Debt']
```

```python
# Strip whitespace from column names.
```

```python
df_gdp.columns.str.strip()
```

```python
df_gdp.columns
```

```python
# Convert GDP_Growth and Inflation columns to numeric (float), removing '%' symbols.
```

```python
df_gdp['GDP_Growth'] = df_gdp['GDP_Growth'].str.rstrip('%').astype(float)
df_gdp['Inflation'] = df_gdp['Inflation'].str.rstrip('%').astype(float)
```

```python
# Time Series Analysis
plt.figure(figsize=(15,5))
sns.lineplot(x='Year', y='GDP_Growth', data=df_gdp)
plt.title('India GDP Growth Rate (1960-2023)')
plt.show()
```

```python
plt.figure(figsize=(18, 8))

df_sector.set_index('Year')[['Exports (in USD billion)', 'Imports (in USD billion)', 'Trade Deficit (in USD billion)']].plot(kind='bar', stacked=True, ax=plt.gca())

plt.title('Exports, Imports, and Trade Deficit (in USD billion)')
plt.xlabel('Year')
plt.ylabel('USD Billion')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Over the years, imports have consistently outpaced exports, resulting in a steady trade deficit.
# Despite occasional variations, the persistent imbalance highlights a heavy reliance on imports in the economy.
```

```python

plt.figure(figsize=(15, 5))
plt.plot(df_gdp['Year'], df_gdp['Inflation'], label='Inflation')
plt.plot(df_gdp['Year'], df_gdp['GDP_Per_Capita'], label='GDP Per Capita')
plt.title('Inflation and GDP Per Capita Over Time')
plt.xlabel('Year')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# From 1990 to 2025, inflation stays low and steady, suggesting stable price levels over the decades.
# Meanwhile, GDP per capita steadily climbs, reflecting significant and sustained economic growth.
```
